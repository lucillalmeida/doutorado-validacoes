```python
# loading the digitized data
case1_data = open("Bed_height_case1_no_distributor.csv", 'r')
time_case1, h_case1 = np.loadtxt(case1_data, delimiter = ',', skiprows = 1, unpack = True)    
case1_data.close()

case2_data = open("Bed_height_case2_no_distributor.csv", 'r')
time_case2, h_case2 = np.loadtxt(case2_data, delimiter = ',', skiprows = 1, unpack = True)    
case2_data.close()

case3_data = open("Bed_height_case3_no_distributor.csv", 'r')
time_case3, h_case3 = np.loadtxt(case3_data, delimiter = ',', skiprows = 1, unpack = True)    
case3_data.close()
```
