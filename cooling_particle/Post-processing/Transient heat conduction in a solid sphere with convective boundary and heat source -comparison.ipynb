# Analytical solutions

## Transient heat conduction in a solid sphere with convective boundary and heat source

Based on Ozisik (2012)
A solid sphere of radius $b$ is initially at temperature $F(r)$. 
For $t >0$, the boundary condition at $r = b$ dissipates heat by convection, with convection coefficient $h$ into a fluid at zero temperature. In addition, the sphere is subjected to uniform internal energy generation $g_0$ (W/m3).

The mathematical formulation is 

\begin{matrix}
\frac{\partial^2 T}{\partial r^2} + \frac{2}{r}\frac{\partial T}{\partial r} + \frac{g_0}{k}= \frac{1}{\alpha} \frac{\partial T}{\partial t}  & \text{in} & 0\leq r\leq b & t>0
\end{matrix}


\begin{matrix}
BC1: \frac{\partial T}{\partial r} (r\rightarrow 0) = 0 \\
BC2: -k \left. \frac{\partial T}{\partial r} \right|_{r=b} = h \left. T \right|_{r=b} \\
IC: T(t=0) = F(r)
\end{matrix}

Introducing a new dependent variable $U(r,T)$, given that

$$ U(r,t) = rT(r,t)$$ 

the equation is now

\begin{matrix}
\frac{\partial^2 U }{\partial r^2} + \frac{g_0 r}{k}= \frac{1}{\alpha} \frac{\partial U}{\partial t}  & \text{in} & 0\leq r\leq b & t>0
\end{matrix}

\begin{matrix}
BC1: U(r = 0) = 0 \\
BC2: \left. \frac{\partial U}{\partial r} \right|_{r=b}  + \left( \frac{h}{k} -  \frac{1}{b} \right)  \left.U \right|_{r=b}  = 0\\
IC: U(t=0) = r F(r)
\end{matrix}


The PDE of equation above is nonhomogeneous; hence we must seek superposition of the form 

$$U(r, t) = \Psi(r, t) + \Phi(r)$$

where $\Psi(r, t)$ takes the homogeneous form of the PDE and boundary conditions, while $\Phi(r)$ is a nonhomogeneous ODE.

### Solution of the nonhomogeneous problem
We first consider the 1-D problem of $\Phi(r)$:

\begin{matrix}
\frac{\partial^2 \Phi }{\partial r^2} + \frac{g_0 r}{k}= 0  & \text{in} & 0\leq r\leq b 
\end{matrix}

\begin{matrix}
BC1: \Phi (r = 0) = 0 \\
BC2: \left. \frac{\partial \Phi}{\partial r} \right|_{r=b} + K \left. \Phi \right|_{r=b} = 0\\
\end{matrix}

where the constant K is introduced:
$$ K = \frac{h}{k}-\frac{1}{b}$$

The solution of equation abovetakes the form of a homogeneous solution plus a particulas solution, giving:

$$\Phi(r) = C_1 + C_2 r - \frac{g_0}{6k}r^3 $$

Boundary condition BC1 eliminates constant $C_1$, while BC2 yields $C_2$:

$$C_2 = \left( \frac{b g_0}{6h} \right) \left(2+ \frac{hb}{k} \right)$$

which gives the final expression for $\Phi(r)$ as

$$\Phi(r) = r \left( C_2 -  \frac{g_0}{6k} r^2 \right) $$

### Solution of the homogeneous problem

We now define the homogeneous $\Psi(r, t)$ problem, noting that $\Phi(r)$ is coupled to the initial condition. The formulation becomes

\begin{matrix}
\frac{\partial^2 \Psi}{\partial r^2} = \frac{1}{\alpha} \frac{\partial \Psi}{\partial t}  & \text{in} & 0\leq r\leq b & t>0
\end{matrix}


\begin{matrix}
BC1: \Psi(r=0) = 0 \\
BC2: \left. \frac{\partial \Psi}{\partial r} \right|_{r=b} + K  \left. \Psi \right|_{r=b} = 0 \\
IC: \Psi(t=0) = rF(r) - \Phi(r)
\end{matrix}

The transient problem is now readily solved with separation of variables given all the homogenous boundary conditions and homogeneous PDE. 

$$\Psi(r,t ) =R(r) \Gamma (t)  $$

which leads to
$$
\frac{1}{R}\frac{d^2 R }{d r^2} = \frac{1}{\alpha \Gamma} \frac{d \Gamma}{d t} = -\lambda^2
$$
Solution of separated ODE in the t dimension yields
$$ \Gamma (t) = C_1 e^{-\alpha \lambda^2 t}$$

while solution of the separated ODE in the r dimension gives
$$ R(r) = C_2 cos \lambda r + C_3 sin \lambda r$$

Boundary condition BC1 yields $C_2 = 0$, while BC2 yields the following transcendental equation

\begin{matrix} 
\lambda_n cot \lambda_n b = -K & \rightarrow & \lambda_n & \text{for} & n=1,2,3...
\end{matrix}


The transcendental equation may be recast in the form $b \lambda_n  cot \lambda_n b + bK =0 $, which has all real roots if $bK >−1$, and noting that $\lambda_n  = 0$ is an eigenvalue only for the special case of $bK = −1$. 
When the value of $K$ as defined above is introduced into the inequality $bK >−1$, we find the necessary condition for all
real roots given as $bh/k >0$ (i.e., $Bi >0$). This result is satisfied with the physical requirements of a positive radius, convection coefficient, and thermal conductivity.

We now sum over all possible solutions, yielding the general solution:

$$ \Psi(r,t) = \sum_{n=1}^\infty C_n sin(\lambda_n r)e^{-\alpha \lambda^2_n t}$$

with $C_n = C_1 C_3 $. The initial condition is now applied, yielding

$$ \Psi(t=0) = r F(r) - \Phi(r)= \sum_{n=1}^\infty C_n sin(\lambda_n r) $$

We solve for the Fourier coefficients by applying the operator

$$\ast\int_{r=0}^b sin \left( \lambda_q r\right) dr $$

which yields the result

$$ C_n = \frac{\int_{r=0}^b \left[ r F(r) - \Phi(r) \right] sin \left( \lambda_n r\right) dr }{\int_{r=0}^b sin^2 \left( \lambda_n r\right) dr}$$

The denominator is the norm $N(\lambda_n)$ and may be simplified with the replacement of $H_2$ with $K$. We finally transform the problem back to $T(r,t )$, yielding the result

$$ T(r,t) = \frac{\Psi(r,t)}{r} + \frac{\Phi(r)}{r} = \sum_{n=1}^\infty C_n \frac{sin(\lambda_n r)}{r} e^{-\alpha \lambda^2_n t} +  \left( C_2 -  \frac{g_0}{6k} r^2 \right) $$

The rightmost term is the steady-state solution, giving the parabolic temperature distribution expected for a steady-state, 1-D sphere with generation.

The average temperature is given as:
$$T_{av}(t) = \frac{\int_0^R 4 \pi r^2 T(r,t) dr}{\frac{4}{3} \pi R^3} = \frac{3}{R^3} \int_0^R r^2 T(r,t) dr \tag{6}$$



```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.optimize import root
from scipy.optimize import fsolve
import scipy.integrate as integrate

def func_cos(x, c):
    return np.cos(x) / x - c

#crange = range(1, 11)

#res = [root(func_cos, 0.5, args=(ci, )).x[0] for ci in crange]
#print(res)

def steadySolution(b, g0, h, k, r):
    C2 = (b*g0/(6*h))*(2+(h*b)/k)
    return C2- g0/(6*k)*r*r

def transcedental_equation(x):
    return x *(1/math.tan(x*b)) + K

def find_lambdas(transcedental_equation, k):
    nrange = range(0, k)
    solution = []
    for i in nrange:
        j = root(transcedental_equation, (i-0.5*math.pi)).x[0]
        solution.append(str('{:f}'.format(j)))

    cleanlist = []
    [cleanlist.append(x) for x in solution if (x not in cleanlist and (float(x) > 0))]

    roots = [float(x) for x in cleanlist]
    return(roots)


        
def computeCn(b, Fr, lambda_n, K, g0, h, k):
    solution = []
    for i in range(len(lambda_n)):
        #print(lambda_n[i])
        A = (integrate.quad(lambda r: (r*Fr-r*steadySolution(b, g0, h, k, r))*math.sin(r*lambda_n[i]), 0, b)[0])
        #B = (integrate.quad(lambda r: math.pow((math.sin(r*lambda_n[i])), 2.0), 0, b)[0])
        result = A*(2*(lambda_n[i]*lambda_n[i]+K*K)/(b*(lambda_n[i]*lambda_n[i]+K*K)+K))
        #result = A/B
        #print(A)
        solution.append(result)
    return(solution)

def Temperature(lambda_n, Cn, r, t, b, g0, h, k):
    T = 0
    Temp = 0
    for i in range(len(lambda_n)):
        #print("i = ", i)
        #print("Cn[i] = ", Cn[i])
        T = T + Cn[i]*(1.0/(r+0.0000000001))*(math.sin(lambda_n[i]*r))*math.exp(-1.0*alpha*lambda_n[i]*lambda_n[i]*t)
    Temp = T+steadySolution(b, g0, h, k, r)
    return(Temp)
    

def average_temperature(lambda_n, Cn, t, R, g0, h, k):
    f = lambda r: r*r*Temperature(lambda_n, Cn, r, t, R, g0, h, k)
    y = integrate.quad(f, 0, R)
    return((3.0/(R*R*R)*y[0]))

def steady_average_temperature(R):
    f = lambda r: r*r*steadySolution(r)
    y = integrate.quad(f, 0, R)
    return((3.0/(R*R*R)*y[0]))
```
